# 11. For all customers aged between 25 - 35, find out:

```python
Customer_Final.head(2)
```

```python
f=pd.to_datetime(dt.datetime.today())
f
```

```python
Customer_Final.DOB
```

```python
Customer_Final['DOB'] = pd.to_datetime(Customer_Final.DOB)
```

```python
Customer_Final['Age']=round((f-Customer_Final.DOB).dt.days/365.25).astype('int')
```

```python
Customer_Final.head()
```

```python
age25_35 = Customer_Final[(Customer_Final.Age>25) & (Customer_Final.Age<35)]
```

```python
# Customers aged between 25 to 35
age25_35
```

# a. What was the total amount spent for “Electronics” and “Books” product categories?

```python
age25_35.prod_cat.value_counts()
```

```python
h=age25_35[(age25_35.prod_cat=='Electronics')|(age25_35.prod_cat=='Books')]
```

```python
#Total amount spent on the 2 categories:

pd.DataFrame(round(h.groupby(['prod_cat'])['total_amt'].sum()))
```

# b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?

```python
(age25_35.tran_date>pd.to_datetime('2014-1-1',format='%Y-%m-%d'))
```

```python
x=pd.to_datetime('2014-1-1',format='%Y-%m-%d')
y=pd.to_datetime('2014-3-1',format='%Y-%m-%d')
```

```python
# Total amount spent by the people aged between 25 to 35 -- Method 1

age25_35[((age25_35.tran_date>x) & (age25_35.tran_date<y))].total_amt.sum()
```

```python
# Total amount spent by the people aged between 25 to 35 -- Method 2

age25_35[((age25_35.tran_date>x) & (age25_35.tran_date<y))]['total_amt'].sum()
```
